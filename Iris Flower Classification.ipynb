```python
import pandas as pd
df=pd.read_csv("E:/iris.csv",names=['sl','sw','pl','pw','class'])
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['class']=le.fit_transform(df['class'])
# separate  input and output
X=df.drop(columns=['class'])
Y=df['class']
# spilitt the dataset for training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=11)
##################

def knn():
    global acc_knn
    global K
    from sklearn.neighbors  import KNeighborsClassifier
    K=KNeighborsClassifier(n_neighbors=5)
    # train the model  using traing dataset(80%)
    K.fit(X_train,Y_train)
    #test the model  by using X_test(10% input)
    Y_pred_knn=K.predict(X_test)
    Y_pred_knn  #  predicted output of X_test
    # find accuracy
    from sklearn.metrics  import accuracy_score
    acc_knn=accuracy_score(Y_test,Y_pred_knn)
    acc_knn=round(acc_knn*100,2)
    m.showinfo(title="KNN",message="accuracy is "+str(acc_knn))
def lg():
    global acc_lg
    from sklearn.linear_model import LogisticRegression
    L=LogisticRegression(solver='liblinear')
    #train the model
    L.fit(X_train,Y_train)
    #test the model
    Y_pred_lg=L.predict(X_test)
    # find accuracy
    from sklearn.metrics import accuracy_score
    acc_lg=accuracy_score(Y_test,Y_pred_lg)
    acc_lg=round(acc_lg*100,2)
    m.showinfo(title="LG",message="accuracy is "+str(acc_lg))
def dt():
    global acc_dt
    from sklearn.tree import DecisionTreeClassifier
    D=DecisionTreeClassifier()
    D.fit(X_train,Y_train)
    Y_pred_dt=D.predict(X_test)
    from sklearn.metrics import accuracy_score
    acc_dt=accuracy_score(Y_test,Y_pred_dt)
    acc_dt=round(acc_dt*100,2)
    m.showinfo(title="DT",message="accuracy is "+str(acc_dt))
def nb():
    global acc_nb
    from sklearn.naive_bayes  import GaussianNB
    N=GaussianNB()
    N.fit(X_train,Y_train)
    Y_pred_nb=N.predict(X_test)
    from sklearn.metrics import accuracy_score
    acc_nb=accuracy_score(Y_test,Y_pred_nb)
    acc_nb=round(acc_nb*100,2)
    m.showinfo(title="NB",message="accuracy is "+str(acc_nb))
    
def cmp():
    import matplotlib.pyplot as plt
    model=['knn','lg','dt','nb']
    acc=[acc_knn,acc_lg,acc_dt,acc_nb]
    plt.bar(model,acc,color=['yellow','green','red','blue'])
    plt.show()
    
def submit():
    sl=float(Esl.get())
    sw=float(Esw.get())
    pl=float(Epl.get())
    pw=float(Epw.get())
    print(sl,sw,pl,pw)
    result=K.predict([[sl,sw,pl,pw]])
    print(result)
    if result[0]==0:
        flower="SEROSA"
    elif  result[0]==1:
        flower ='VERSICOLOR'
    else:
        flower="VIRGINICA"
    m.showinfo(title="result",message=flower)    
def reset():
    Esl.delete(0,END)
    Esw.delete(0,END)
    Epl.delete(0,END)
    Epw.delete(0,END)
    
    
    
    
from tkinter import *
import tkinter.messagebox as m
w=Tk()
#  first row
L1=Label(w,text="IRIS FLOWER CLASSIFICATION",font=('arial',15,'bold'))
L1.grid(row=1,column=1,columnspan=4)
#2nd row
Bknn=Button(w,text="KNN",font=('arial',15,'bold'),command=knn)
Blg=Button(w,text="LG",font=('arial',15,'bold'),command=lg)
Bdt=Button(w,text="DT",font=('arial',15,'bold'),command=dt)
Bnb=Button(w,text="NB",font=('arial',15,'bold'),command=nb)
Bknn.grid(row=2,column=1)
Blg.grid(row=2,column=2)
Bdt.grid(row=2,column=3)
Bnb.grid(row=2,column=4)
#3rd  row
Bcmp=Button(w,text="Compare",font=('arial',15,'bold'),command=cmp)
Bcmp.grid(row=3,column=2,columnspan=2)

#4th row  
L2=Label(w,text="Predict  for a new flower",font=('arial',15,'bold'))
L2.grid(row=4,column=1,columnspan=4)

# 5th  tow
Lsl=Label(w,text="SL",font=('arial',15,'bold'))
Esl=Entry(w,font=('arial',15,'bold'),width=10)
Lsw=Label(w,text="SW",font=('arial',15,'bold'))
Esw=Entry(w,font=('arial',15,'bold'),width=10)

Lsl.grid(row=5,column=1)
Esl.grid(row=5,column=2)
Lsw.grid(row=5,column=3)
Esw.grid(row=5,column=4)
#6th row
Lpl=Label(w,text="PL",font=('arial',15,'bold'))
Epl=Entry(w,font=('arial',15,'bold'),width=10)
Lpw=Label(w,text="PW",font=('arial',15,'bold'))
Epw=Entry(w,font=('arial',15,'bold'),width=10)

Lpl.grid(row=6,column=1)
Epl.grid(row=6,column=2)
Lpw.grid(row=6,column=3)
Epw.grid(row=6,column=4)

# 7 th row 
Bsub=Button(w,text="submit",font=('arial',15,'bold'),command=submit)
Bres=Button(w,text="reset",font=('arial',15,'bold'),command=reset)
Bsub.grid(row=7,column=1,columnspan=2)
Bres.grid(row=7,column=3,columnspan=2)
w.mainloop()
```
